```python
# Verificar possíveis valores inválidos nas principais colunas numéricas
valores_invalidos = {
    "quantidade <= 0": (df["quantidade"] <= 0).sum(),
    "preco_unitario <= 0": (df["preco_unitario"] <= 0).sum(),
    "desconto < 0": (df["desconto"] < 0).sum(),
    "desconto > 1": (df["desconto"] > 1).sum(),
    "custo_unitario <= 0": (df["custo_unitario"] <= 0).sum(),
    "receita < 0": (df["receita"] < 0).sum(),
    "lucro < 0": (df["lucro"] < 0).sum(),
    "meta_venda <= 0": (df["meta_venda"] <= 0).sum(),
    "prazo_entrega_dias < 0": (df["prazo_entrega_dias"] < 0).sum()
}

valores_invalidos = pd.DataFrame(
    valores_invalidos.items(),
    columns=["verificacao", "qtd_registros"]
)

valores_invalidos
```

```python
# Verificar consistência da receita
# Receita esperada = quantidade * preco_unitario * (1 - desconto)

receita_calculada = df["quantidade"] * df["preco_unitario"] * (1 - df["desconto"])

# Diferença absoluta entre receita da base e receita calculada
diferenca_receita = (df["receita"] - receita_calculada).abs()

# Considerar inconsistência quando a diferença for maior que 0,01
inconsistencias_receita = df[diferenca_receita > 0.01]

print(f"Quantidade de inconsistências na receita: {len(inconsistencias_receita)}")

inconsistencias_receita[
    [
        "id_venda",
        "produto",
        "quantidade",
        "preco_unitario",
        "desconto",
        "receita"
    ]
]
```

```python
# Verificar consistência do lucro
# Lucro esperado = receita - (quantidade * custo_unitario)

lucro_calculado = df["receita"] - (df["quantidade"] * df["custo_unitario"])

# Diferença absoluta entre lucro da base e lucro calculado
diferenca_lucro = (df["lucro"] - lucro_calculado).abs()

# Considerar inconsistência quando a diferença for maior que 0,01
inconsistencias_lucro = df[diferenca_lucro > 0.01]

print(f"Quantidade de inconsistências no lucro: {len(inconsistencias_lucro)}")

inconsistencias_lucro[
    [
        "id_venda",
        "produto",
        "quantidade",
        "receita",
        "custo_unitario",
        "lucro"
    ]
]
```

```python
# Verificar consistência da coluna atingiu_meta
# Se receita >= meta_venda, então deveria ser "Sim"; caso contrário, "Não"

atingiu_meta_calculado = np.where(df["receita"] >= df["meta_venda"], "Sim", "Não")

inconsistencias_meta = (df["atingiu_meta"] != atingiu_meta_calculado).sum()

print(f"Inconsistências na coluna atingiu_meta: {inconsistencias_meta}")
```

```python
# Selecionar colunas numéricas relevantes para análise de outliers
colunas_numericas_analise = [
    "tempo_relacionamento_meses",
    "quantidade",
    "preco_unitario",
    "desconto",
    "receita",
    "custo_unitario",
    "lucro",
    "meta_venda",
    "prazo_entrega_dias"
]

resultado_outliers = []

for coluna in colunas_numericas_analise:
    q1 = df[coluna].quantile(0.25)
    q3 = df[coluna].quantile(0.75)
    iqr = q3 - q1
    
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr
    
    qtd_outliers = ((df[coluna] < limite_inferior) | (df[coluna] > limite_superior)).sum()
    
    resultado_outliers.append({
        "coluna": coluna,
        "q1": q1,
        "q3": q3,
        "limite_inferior": limite_inferior,
        "limite_superior": limite_superior,
        "qtd_outliers": qtd_outliers
    })

outliers_df = pd.DataFrame(resultado_outliers)

outliers_df
```

# Relatório de Diagnóstico e Compreensão de Dados

## 2. Etapa 2 - Compreensão e Diagnóstico dos Dados

A fase de diagnóstico é fundamental para garantir a integridade das análises subsequentes. A base de dados da **Comercial Insight Ltda.** compreende 1.020 registros e 27 atributos, cobrindo o ciclo operacional de vendas, clientes, logística e performance comercial no período de **01/01/2024 a 30/12/2024**.

### 2.1. Diagnóstico de Qualidade dos Dados (Data Quality)

Após uma auditoria inicial na base, foram identificados pontos críticos que exigem intervenção na etapa de pré-processamento:

* **Valores Ausentes (Missing Data):** Identificamos 51 registros (5% da base) com vacância nas colunas críticas: `faixa_renda`, `preco_unitario`, `desconto` e `custo_unitario`. A ausência ocorre de forma simultânea nessas linhas, o que aponta para uma falha sistêmica na integração de dados ou no processo de exportação do ERP/CRM.
* **Redundância e Duplicidade:** Foram detectadas 20 linhas integralmente duplicadas e 20 duplicidades específicas na chave primária `id_venda`. Considerando que a base deveria conter 1.000 transações únicas, a remoção desses registros é mandatória para evitar a inflação artificial dos indicadores financeiros.
* **Inconsistências Lógicas de Negócio:** * **Quantidade Zero:** Foram localizados 10 registros com `quantidade` igual a zero, porém apresentando `receita` e `lucro` positivos. Este é um erro grave de integridade lógica, invalidando os cálculos de preço médio e volume.
    * **Erros de Classificação Regional:** 30 registros foram classificados na região `Exterior`, apesar de as cidades listadas serem brasileiras. Isso exigirá um saneamento geográfico para não distorcer o desempenho regional.
* **Integridade Temporal e Financeira:** * As colunas de granularidade temporal (`mes` e `trimestre`) estão íntegras e coerentes com a data da venda.
    * Foram identificados 126 registros com **lucro negativo**. Diferente dos outros erros, estes casos serão mantidos para análise, pois representam a realidade operacional (vendas com prejuízo ou queima de estoque), exceto onde houver conflito com as inconsistências de quantidade zero.

### 2.2. Identificação e Descrição dos Atributos

Dada a ausência de um dicionário de dados formal, realizamos a engenharia reversa dos metadados. Os 27 atributos foram mapeados para cobrir quatro pilares principais: **Transacional** (venda, data, valores), **Entidades** (cliente, vendedor), **Logística** (entrega, prazos) e **Estratégico** (metas, atingimento).

Esta classificação é essencial para a estruturação do futuro Modelo Dimensional (Star Schema), onde separaremos o que é métrica (Fato) do que é contexto (Dimensões).

### 2.3. Classificação Técnica das Variáveis

A classificação abaixo orientará o tratamento estatístico e a escolha dos algoritmos de visualização:

| Categoria | Tipo de Dado | Atributos |
| :--- | :--- | :--- |
| **Identificadores** | Qualitativo Nominal | `id_venda`, `cliente` |
| **Temporais** | Time Series | `data`, `mes`, `trimestre` |
| **Categóricas Nominais** | Qualitativo Nominal | `tipo_cliente`, `segmento`, `produto`, `categoria`, `marca`, `regiao`, `cidade`, `canal_venda`, `vendedor`, `status_entrega` |
| **Categóricas Ordinais** | Qualitativo Ordinal | `faixa_renda`, `linha_produto`, `atingiu_meta` |
| **Numéricas Discretas** | Quantitativo Discreto | `quantidade`, `tempo_relacionamento_meses`, `prazo_entrega_dias` |
| **Numéricas Contínuas** | Quantitativo Contínuo | `preco_unitario`, `desconto`, `receita`, `custo_unitario`, `lucro`, `meta_venda` |

**Nota Técnica:** A variável `data`, embora lida inicialmente como *string* (objeto), será tipada como `datetime64` na Etapa 3 para permitir cálculos de sazonalidade e séries temporais.
