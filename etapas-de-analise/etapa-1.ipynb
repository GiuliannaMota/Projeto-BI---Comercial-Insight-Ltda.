```python
# ETAPA 1 - ENTENDIMENTO DO PROBLEMA DE NEGÓCIO

# Importação das bibliotecas principais
import pandas as pd
import numpy as np

# Configuração para melhorar a visualização das tabelas no Jupyter
pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:,.2f}".format)

# Caminho do arquivo CSV
caminho_arquivo = "dados-empresa.csv"

# Carregamento da base de dados
df = pd.read_csv(caminho_arquivo)

# Quantidade de linhas e colunas
linhas, colunas = df.shape

print(f"A base possui {linhas} linhas e {colunas} colunas.")

# Listar os nomes das colunas da base
df.columns.tolist()
```

```python
# KPIs iniciais da base bruta
# ============================================================

receita_total = df["receita"].sum()
lucro_total = df["lucro"].sum()
margem_lucro = lucro_total / receita_total
quantidade_total = df["quantidade"].sum()
ticket_medio = df["receita"].mean()
desconto_medio = df["desconto"].mean()

taxa_atingimento_meta = (df["atingiu_meta"] == "Sim").mean()
taxa_atraso_entrega = (df["status_entrega"] == "Atrasado").mean()

kpis_iniciais = pd.DataFrame({
    "KPI": [
        "Receita Total",
        "Lucro Total",
        "Margem de Lucro",
        "Quantidade Vendida",
        "Ticket Médio por Venda",
        "Desconto Médio",
        "Taxa de Atingimento de Meta",
        "Taxa de Entregas Atrasadas"
    ],
    "Valor": [
        receita_total,
        lucro_total,
        margem_lucro,
        quantidade_total,
        ticket_medio,
        desconto_medio,
        taxa_atingimento_meta,
        taxa_atraso_entrega
    ]
})

kpis_iniciais
```

## Etapa 1 - Entendimento do Problema de Negócio

A Comercial Insight Ltda. é uma empresa fictícia do setor de vendas de produtos de tecnologia, atuando em diferentes regiões e utilizando canais físicos e digitais. A organização comercializa produtos como notebooks, smartphones, tablets, monitores e periféricos, atendendo clientes de diferentes perfis, segmentos e faixas de renda.

O principal desafio da empresa é compreender as variações no desempenho de vendas, identificar regiões e produtos com baixa margem de lucro, avaliar diferenças entre canais de venda e entender o comportamento dos clientes. Para isso, será desenvolvido um projeto de Business Intelligence com foco em apoiar a tomada de decisão baseada em dados.

O problema de decisão definido para este projeto é: como a Comercial Insight Ltda. pode utilizar seus dados de vendas para identificar quais produtos, regiões, canais, clientes e vendedores mais contribuem ou prejudicam o desempenho financeiro da empresa, permitindo decisões estratégicas para aumentar receita, lucro, margem e eficiência comercial?

As perguntas de negócio que orientarão a análise são:

1. Quais regiões geram maior receita e quais apresentam menor margem de lucro?
2. Quais produtos e categorias são mais vendidos e quais geram maior lucro?
3. O canal online apresenta desempenho melhor ou pior que o canal loja em termos de receita, lucro e atingimento de metas?
4. Quais vendedores possuem melhor desempenho em receita, lucro e cumprimento de metas?
5. Clientes recorrentes geram maior receita e lucro do que clientes novos?
6. O nível de desconto concedido impacta negativamente a margem de lucro?
7. Existe relação entre atraso na entrega e pior desempenho comercial?
8. Quais marcas ou linhas de produto apresentam maior contribuição para o lucro total?

Os principais KPIs definidos para acompanhar o desempenho da empresa são: receita total, lucro total, margem de lucro, quantidade vendida, ticket médio por venda, desconto médio, taxa de atingimento de meta, taxa de entregas atrasadas, lucro por região e receita por canal de venda.
