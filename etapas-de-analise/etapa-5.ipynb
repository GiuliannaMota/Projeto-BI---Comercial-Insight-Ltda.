# Estrutura Analítica  -  Modelo Dimensional

### Tabela Fato
Armazena os eventos principais do negócio. Neste caso, cada linha representa uma venda realizada.  
Ela contém:
- chaves para as dimensões;
- medidas numéricas (quantidade, preço unitário, desconto, receita, custo unitário, lucro, meta de venda, prazo de entrega);
- indicadores de desempenho.

### Tabelas Dimensão
As dimensões descrevem o contexto da venda.  
|Dimensão|Responde perguntas como|
|--|--|
|Tempo|Quando a venda aconteceu|
|Cliente|Quem comprou? Qual perfil|
|Produto|O que foi vendido?|
|Geografia|Onde foi vendido?|
|Canal|Por qual canal foi vendido|
|Vendedor|Quem vendeu?|
|Entrega|Entregue no prazo ou atrasou?|

### Grão Tabela
O grão é o nível de detalhe da tabela fato. Nesse caso será: **Uma linha da tabela fato representa uma venda individual realizada pela empresa.**  
Ou seja: *1 linha = 1 venda = 1id_venda*



```python
# ETAPA 5 - MODELAGEM PARA BI
# MODELO DIMENSIONAL EM ESQUEMA ESTRELA
# ============================================================

import pandas as pd
import numpy as np

# Configurações de visualização
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.float_format", "{:,.2f}".format)

# Carregar a base tratada gerada na Etapa 3
df = pd.read_csv("dados_empresa_tratado.csv")

# Converter a coluna data para datetime
df["data"] = pd.to_datetime(df["data"], errors="coerce")

# Verificar estrutura
print(df.to_string())
```

```python
# DIMENSÃO TEMPO
# ============================================================

dim_tempo = df[["data"]].drop_duplicates().copy()

# Criar chave substituta baseada na data no formato AAAAMMDD
dim_tempo["id_tempo"] = dim_tempo["data"].dt.strftime("%Y%m%d").astype(int)

# Criar atributos temporais
dim_tempo["ano"] = dim_tempo["data"].dt.year
dim_tempo["mes"] = dim_tempo["data"].dt.month
dim_tempo["nome_mes"] = dim_tempo["data"].dt.month_name()
dim_tempo["trimestre"] = dim_tempo["data"].dt.quarter
dim_tempo["semestre"] = np.where(dim_tempo["mes"] <= 6, 1, 2)
dim_tempo["dia"] = dim_tempo["data"].dt.day
dim_tempo["dia_semana"] = dim_tempo["data"].dt.day_name()

# Reordenar colunas
dim_tempo = dim_tempo[
    [
        "id_tempo",
        "data",
        "ano",
        "mes",
        "nome_mes",
        "trimestre",
        "semestre",
        "dia",
        "dia_semana"
    ]
].sort_values("data").reset_index(drop=True)

dim_tempo
```

```python
# DIMENSÃO CLIENTE
# ============================================================

dim_cliente = df[
    [
        "cliente",
        "tipo_cliente",
        "tempo_relacionamento_meses",
        "faixa_renda",
        "segmento"
    ]
].drop_duplicates().copy()

# Criar chave substituta sequencial
dim_cliente = dim_cliente.reset_index(drop=True)
dim_cliente["id_cliente"] = dim_cliente.index + 1

# Reordenar colunas
dim_cliente = dim_cliente[
    [
        "id_cliente",
        "cliente",
        "tipo_cliente",
        "tempo_relacionamento_meses",
        "faixa_renda",
        "segmento"
    ]
]

dim_cliente
```

```python
# DIMENSÃO PRODUTO
# ============================================================

dim_produto = df[
    [
        "produto",
        "categoria",
        "marca",
        "linha_produto"
    ]
].drop_duplicates().copy()

# Criar chave substituta sequencial
dim_produto = dim_produto.reset_index(drop=True)
dim_produto["id_produto"] = dim_produto.index + 1

# Reordenar colunas
dim_produto = dim_produto[
    [
        "id_produto",
        "produto",
        "categoria",
        "marca",
        "linha_produto"
    ]
]

dim_produto
```

```python
# DIMENSÃO GEOGRAFIA
# ============================================================

dim_geografia = df[
    [
        "cidade",
        "regiao"
    ]
].drop_duplicates().copy()

# Criar chave substituta sequencial
dim_geografia = dim_geografia.reset_index(drop=True)
dim_geografia["id_geografia"] = dim_geografia.index + 1

# Reordenar colunas
dim_geografia = dim_geografia[
    [
        "id_geografia",
        "cidade",
        "regiao"
    ]
]

dim_geografia 
```

```python
# DIMENSÃO CANAL
# ============================================================

dim_canal = df[["canal_venda"]].drop_duplicates().copy()

# Criar chave substituta sequencial
dim_canal = dim_canal.reset_index(drop=True)
dim_canal["id_canal"] = dim_canal.index + 1

# Reordenar colunas
dim_canal = dim_canal[
    [
        "id_canal",
        "canal_venda"
    ]
]

dim_canal
```

```python
# DIMENSÃO VENDEDOR
# ============================================================

dim_vendedor = df[["vendedor"]].drop_duplicates().copy()

# Criar chave substituta sequencial
dim_vendedor = dim_vendedor.reset_index(drop=True)
dim_vendedor["id_vendedor"] = dim_vendedor.index + 1

# Reordenar colunas
dim_vendedor = dim_vendedor[
    [
        "id_vendedor",
        "vendedor"
    ]
]

dim_vendedor
```

```python
# DIMENSÃO ENTREGA
# ============================================================

dim_entrega = df[["status_entrega"]].drop_duplicates().copy()

# Criar chave substituta sequencial
dim_entrega = dim_entrega.reset_index(drop=True)
dim_entrega["id_entrega"] = dim_entrega.index + 1

# Reordenar colunas
dim_entrega = dim_entrega[
    [
        "id_entrega",
        "status_entrega"
    ]
]

dim_entrega
```

```python
# CRIAÇÃO DA TABELA FATO
# ============================================================

# Começar com uma cópia da base tratada
fato_vendas = df.copy()

# Criar id_tempo na tabela fato
fato_vendas["id_tempo"] = fato_vendas["data"].dt.strftime("%Y%m%d").astype(int)

# Adicionar id_cliente
fato_vendas = fato_vendas.merge(
    dim_cliente,
    on=[
        "cliente",
        "tipo_cliente",
        "tempo_relacionamento_meses",
        "faixa_renda",
        "segmento"
    ],
    how="left"
)

# Adicionar id_produto
fato_vendas = fato_vendas.merge(
    dim_produto,
    on=[
        "produto",
        "categoria",
        "marca",
        "linha_produto"
    ],
    how="left"
)

# Adicionar id_geografia
fato_vendas = fato_vendas.merge(
    dim_geografia,
    on=[
        "cidade",
        "regiao"
    ],
    how="left"
)

# Adicionar id_canal
fato_vendas = fato_vendas.merge(
    dim_canal,
    on="canal_venda",
    how="left"
)

# Adicionar id_vendedor
fato_vendas = fato_vendas.merge(
    dim_vendedor,
    on="vendedor",
    how="left"
)

# Adicionar id_entrega
fato_vendas = fato_vendas.merge(
    dim_entrega,
    on="status_entrega",
    how="left"
)

# Criar campo binário para atingimento de meta
fato_vendas["atingiu_meta_flag"] = np.where(
    fato_vendas["atingiu_meta"] == "Sim",
    1,
    0
)

# Selecionar apenas chaves e medidas para a tabela fato
fato_vendas = fato_vendas[
    [
        "id_venda",
        "id_tempo",
        "id_cliente",
        "id_produto",
        "id_geografia",
        "id_canal",
        "id_vendedor",
        "id_entrega",
        "quantidade",
        "preco_unitario",
        "desconto",
        "receita",
        "custo_unitario",
        "lucro",
        "meta_venda",
        "atingiu_meta_flag",
        "prazo_entrega_dias"
    ]
]

fato_vendas
```

```python
# RESUMO DAS TABELAS DO MODELO DIMENSIONAL
# ============================================================

resumo_modelo = pd.DataFrame({
    "tabela": [
        "fato_vendas",
        "dim_tempo",
        "dim_cliente",
        "dim_produto",
        "dim_geografia",
        "dim_canal",
        "dim_vendedor",
        "dim_entrega"
    ],
    "linhas": [
        fato_vendas.shape[0],
        dim_tempo.shape[0],
        dim_cliente.shape[0],
        dim_produto.shape[0],
        dim_geografia.shape[0],
        dim_canal.shape[0],
        dim_vendedor.shape[0],
        dim_entrega.shape[0]
    ],
    "colunas": [
        fato_vendas.shape[1],
        dim_tempo.shape[1],
        dim_cliente.shape[1],
        dim_produto.shape[1],
        dim_geografia.shape[1],
        dim_canal.shape[1],
        dim_vendedor.shape[1],
        dim_entrega.shape[1]
    ]
})

resumo_modelo
```

```python
# VERIFICAÇÃO DE CHAVES ESTRANGEIRAS NULAS
# ============================================================

chaves_estrangeiras = [
    "id_tempo",
    "id_cliente",
    "id_produto",
    "id_geografia",
    "id_canal",
    "id_vendedor",
    "id_entrega"
]

fato_vendas[chaves_estrangeiras].isna().sum()
```

```python
# VERIFICAÇÃO DA CHAVE PRIMÁRIA DA TABELA FATO
# ============================================================

print("Quantidade de linhas na fato:", fato_vendas.shape[0])
print("Quantidade de id_venda únicos:", fato_vendas["id_venda"].nunique())
print("IDs duplicados:", fato_vendas["id_venda"].duplicated().sum())
```

```python
# CONFERÊNCIA DOS TOTAIS DA BASE ORIGINAL TRATADA E DA FATO
# ============================================================

comparacao_totais = pd.DataFrame({
    "indicador": [
        "Receita Total",
        "Lucro Total",
        "Quantidade Vendida",
        "Meta Total"
    ],
    "base_tratada": [
        df["receita"].sum(),
        df["lucro"].sum(),
        df["quantidade"].sum(),
        df["meta_venda"].sum()
    ],
    "fato_vendas": [
        fato_vendas["receita"].sum(),
        fato_vendas["lucro"].sum(),
        fato_vendas["quantidade"].sum(),
        fato_vendas["meta_venda"].sum()
    ]
})

comparacao_totais["diferenca"] = (
    comparacao_totais["base_tratada"] - comparacao_totais["fato_vendas"]
)

comparacao_totais
```

```python
# EXPORTAR TABELAS DO MODELO DIMENSIONAL
# ============================================================

fato_vendas.to_csv("fato_vendas.csv", index=False, encoding="utf-8-sig")
dim_tempo.to_csv("dim_tempo.csv", index=False, encoding="utf-8-sig")
dim_cliente.to_csv("dim_cliente.csv", index=False, encoding="utf-8-sig")
dim_produto.to_csv("dim_produto.csv", index=False, encoding="utf-8-sig")
dim_geografia.to_csv("dim_geografia.csv", index=False, encoding="utf-8-sig")
dim_canal.to_csv("dim_canal.csv", index=False, encoding="utf-8-sig")
dim_vendedor.to_csv("dim_vendedor.csv", index=False, encoding="utf-8-sig")
dim_entrega.to_csv("dim_entrega.csv", index=False, encoding="utf-8-sig")

print("Tabelas do modelo dimensional exportadas com sucesso.")
```

### Tabela Fato (Fato_Vendas)
**Descrição:** Entidade central que armazena os eventos quantificáveis de cada transação de venda e operação logística da empresa.  
  
**Chaves Estrangeiras (FK) para cruzamento:**
- id_tempo = conecta com Dim_Tempo
- id_produto = Conecta com a Dim_Produto
- id_cliente = Conecta com a Dim_Cliente
- id_geografia = Conecta com a Dim_Geografia
- id_vendedor = Conecta com a Dim_Vendedor
- id_canal = Conecta com a Dim_Canal
- id_entrega = Conecta com a Dim_Entrega

**Métricas de Fato (campos de cálculo analítico):**  
quantidade, preco_unitario, custo_unitario, desconto, receita, lucro, prazo_entrega_dias

## Análise e Definiçao das Dimensões
Mapeamento detalhado de como cada dimensão.

1. **Dimensão Produto (Dim_Produto)**  
- Campos analisados: id_produto, produto, categoria, marca, linha_produto.
- Propósito: permitir a quebra de receita e lucro por níveis hierárquicos. Com ela, pode-se responder quais mercadorias são mais vendidas e o peso das marcas (Alpha, Beta, Gamma, Delta) ou linhas (Econômico, Padrão, Premium) no lucro final.

2. **Dimensão Geografia (Dim_Geografia)**
- Campos analisados: id_geografia, cidade, regiao.
- Propósito: fornecer o contexto espacial do negócio. Usada para analisar quais regiões geram o maior faturamento bruto e onde se concentram os piores índices logísticos.

3. **Dimensão Cliente (Dim_Cliente)**
- Campos analisados: id_cliente, cliente, tipo_cliente, tempo_relacionamento_meses, faixa_renda, segmento.
- Propósito: usado para estudo de comportamento e perfil de consumo. Permite analisar se clientes recorrentes compram mais que clientes novos, o impacto do tempo de relacionamento no lucro e o comportamento de canais como Varejo e Atacado.

4. **Dimensão Tempo (Dim_Tempo)**
- Campos analisados: id_tempo, data, ano, mes, nome_mes, trimestre, semestre, dia, dia_semana.
- Propósito: espinha dorsal de qualquer análise de série temporal. Permite calcular o crescimento mês a mês, identificar sazonalidades de dias específicos da semana e alimentar gráficos de tendência de receita diária.

5. **Dimensão Vendedor (Dim_Vendedor)**
- Campos analisados: id_vendedor, vendedor.
- Propósito: zvaliação de performance de recursos humanos. Cruzada com a Fato, ela isola o desempenho individual de cada membro da equipe comercial em faturamento e lucratividade.

6. **Dimensão Canal (Dim_Canal)**
- Campos analisados: id_canal, canal_venda.
- Propósito: avaliação de canais de aquisição. Responde à pergunta estratégica se a operação Online performa melhor ou pior que as Lojas físicas.

7. **Dimensão Entrega (Dim_Entrega)**
- Campos analisados: id_entrega, status_entrega.
- Propósito: suditoria e controle de qualidade operacional (SLA). Usada para cruzar pedidos "No prazo" ou "Atrasado" com as perdas de desempenho comercial.
